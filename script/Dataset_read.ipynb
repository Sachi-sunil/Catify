```python
with open('Spotify Account Data/YourSoundCapsule.json', 'r') as f:
    data = json.load(f)

stats = data["stats"]

# --------------------------------------
# DAILY SUMMARY
# --------------------------------------
df_daily = pd.DataFrame(stats)
display(df_daily)

# --------------------------------------
# TOP TRACKS (with metadata prefix)
# --------------------------------------
df_top_tracks = pd.json_normalize(
    stats,
    record_path="topTracks",
    meta=["date", "streamCount", "secondsPlayed"],
    meta_prefix="day_"
)

display(df_top_tracks)

# --------------------------------------
# TOP ARTISTS (with metadata prefix)
# --------------------------------------
df_top_artists = pd.json_normalize(
    stats,
    record_path="topArtists",
    meta=["date", "streamCount", "secondsPlayed"],
    meta_prefix="day_"
)

display(df_top_artists)

# --------------------------------------
# TOP GENRES
# --------------------------------------
df_top_genres = pd.json_normalize(
    stats,
    record_path="topGenres",
    meta=["date", "streamCount", "secondsPlayed"],
    meta_prefix="day_"
)

display(df_top_genres)
```
